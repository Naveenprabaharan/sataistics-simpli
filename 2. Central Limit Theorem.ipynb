```python
# import packages
from numpy.random import seed
from numpy.random import randint
from numpy import mean
import matplotlib.pyplot as plt
```

# 1. Sample mean

```python
rolls = randint(1,7,100)
rolls
```

```python
rolls.mean()
```

# 2. Population mean

```python
seed(1)
values = [randint(1,7,100).mean() for i in range(1000)]
```

```python
plt.hist(values, bins=10)
plt.show()
```
